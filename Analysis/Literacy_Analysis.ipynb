```python
import csv
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.io as pio

# set default
pio.templates.default = "plotly_dark"
```

```python
# read data
df = pd.read_csv('All_States_GE.csv')
df
```

```python
# Unique Education Bars
ed = [x for x in list(df["MyNeta_education"]) if str(x) != 'nan']
# count
ed_count = {}
for i in ed:
    if i in ed_count:
        ed_count[i] += 1
    else:
        ed_count[i] = 1

# sort
ed_count = {k: v for k, v in sorted(ed_count.items(), key=lambda item: item[1], reverse=True)}
ed_count
```

```python
# Education Map according to documentation
ed_map = {
    "Others": 0,
    "Illiterate": 1,
    "Literate": 2,
    "5th Pass": 3,
    "8th Pass": 4,
    "10th Pass": 5,
    "12th Pass": 6,
    "Graduate": 7,
    "Graduate Professional": 8,
    "Post Graduate": 9,
    "Doctorate": 10
}

stateVals = {}
# iterate through rows
for i in range(len(df)):
    # get education
    row = df.loc[i]
    stateName = row["State_Name"]
    ed = row["MyNeta_education"]
    # check if nan
    if str(ed) == 'nan':
        continue

    # get education value
    edVal = ed_map[ed]
    # check if state exists
    if stateName in stateVals:
        stateVals[stateName].append(edVal)
    else:
        stateVals[stateName] = [edVal]

# sort based on mean
stateVals = {k: v for k, v in sorted(stateVals.items(), key=lambda item: np.mean(item[1]), reverse=True)}

means = {}
for state in stateVals:
    print(f"State: {state}")
    # Mean std dev
    means[state] = np.mean(stateVals[state])
    print(f"Mean: {means[state] }")
    print(f"Std Dev: {np.std(stateVals[state])}")
    
    print()
```

```python
# INDIA State value\n
stateMin = min(means, key=means.get)
stateMax = max(means, key=means.get)
for state in stateVals:
    stateVal = means[state]
    # scale from 0 to 100
    stateVal = (stateVal - means[stateMin]) / (means[stateMax] - means[stateMin]) * 100


    print(f"INDIA\t{state.upper().replace('_', ' ')}\t{stateVal}")
```

```python
binaryVals = {}
for state in stateVals:
    binaryVals[state] = []
    for val in stateVals[state]:
        if val <= 1:
            binaryVals[state].append(0)
        else:
            binaryVals[state].append(1)

binaryMeans = {}
for state in binaryVals:
    binaryMeans[state] = np.mean(binaryVals[state])

binaryMin = min(binaryMeans, key=binaryMeans.get)
binaryMax = max(binaryMeans, key=binaryMeans.get)

for state in binaryVals:
    stateVal = binaryMeans[state]
    # scale from 0 to 100
    stateVal = (stateVal - binaryMeans[binaryMin]) / (binaryMeans[binaryMax] - binaryMeans[binaryMin]) * 100

    print(f"INDIA\t{state.upper().replace('_', ' ')}\t{stateVal}")


```
